```python
generate_canvas( w = .4 * 165 / 24.5, a = 1.5 )

K = -1 
tau_0 = 200e-6 

tau_rhp = 100e-6 

N = 2

def X(, 0, NF):
    N = (NF - 1)/2
    return 1/(2*NF)*(sin( (-0)*(N + 0.5) )/sin( (-0)/2 ) \
                     + \
                     sin( (+0)*(N + 0.5) )/sin( (+0)/2 ) )

_vec  = linspace(-pi, pi, 1000)
_dsk = arange(-5, 6) * pi/5 + 0.001

v1     = X(_vec, pi/5, 9)
v1_dsk = X(_dsk, pi/5, 9)


plot(_vec/pi, abs(v1), 'k--', lw = 0.8)
stem(_dsk/pi, abs(v1_dsk),  use_line_collection=True, basefmt="k", markerfmt="k.", linefmt = 'k')

xlabel(r"$\Omega/\uppi$")
ylabel(r"$X({\rm j}\Omega)$")

#legend([r"$y$", r"$x$"])

xlim(-1, 1)
ylim(-0.2 ,1)

plt.savefig("sample_X_leak.pdf", bbox_inches='tight')
```
