```python
from pylab import *
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif"
    })

plt.rc('text.latex', preamble=r'\usepackage[T1,T2A]{fontenc}\usepackage[utf8]{inputenc}\usepackage{cmsrb}\usepackage{amsmath}\usepackage[Symbolsmallscale]{upgreek}\usepackage{icomma}')

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)


# Decimalni zarez
import locale
locale.setlocale(locale.LC_NUMERIC, "sr_RS.utf8")
plt.rcParams['axes.formatter.use_locale'] = True


def u(t):
    return (t > 0) * 1.0

def generate_canvas(w = 1.6, a = 1.618):
    fig = figure (figsize = (w, w/a), dpi = 200)
    xlim([-2,2])
    ylim([-2,2])
    minorticks_on()
    grid(b=True, which='major', color='0.45', linestyle='-')
    grid(b=True, which='minor', color='0.75', linestyle='--')

    return fig

def plot_dirac(x, A, lw = 1.5, c = 'k'):
    arrow(x, 0, 0, A, shape="full", color=c,linewidth=lw, head_width=0.2, head_length=0.3, length_includes_head=True, zorder=10)
    
    
```

```python
generate_canvas( w = 165 / 24.5, a = 4 )
ylim([-1.5, 1.5])

T = 20

def f(t):
    if t < 0: return 0
    if t < T: return 62.5*sqrt(  (1 - exp(-t/18)) )
    return 62.5 * sqrt ( ( exp( T/18 ) - 1 ) * exp(-t/18) )

def lin(t):
    if t < 0: return 0
    if t < T: return 62.5*(  (1 - exp(-t/18)) )
    return 62.5 * ( ( exp( T/18 ) - 1 ) * exp(-t/18) )


tmax = 80
t = linspace(0, tmax, 1000)

v = [ f(t) for t in t ]

xlim([0, tmax])
ylim([0, 65])

plot(t, v, 'k-', lw = 1.5)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$v\rm\,[km/h]$")

#plt.savefig("tram_plot.pdf", bbox_inches='tight')
```

```python
generate_canvas( w = .65 * 165 / 24.5, a = 2 )
#ylim([-1.5, 1.5])

K = -1 
tau_0 = 200e-6 

tau_rhp = 100e-6 


def f(t):
    if t < 0: return 0
    return K * (1 - exp(-t/tau_0))

def g(t):
    if t < 0: return 0
    return K + (1-K)*exp(-t/tau_rhp)


tmax = 1000e-6
t_vec = linspace(-tmax, tmax, 2000)

v1 = [ f(t) for t in t_vec ]
v2 = [ g(t) for t in t_vec ]


plot(t_vec/1e-3, v1, 'k-', lw = 1.5)
plot(t_vec/1e-3, v2, 'k--', lw = 1)
xlabel(r"$t\rm\,[ms]$")
ylabel(r"$v\rm\,[V]$")

xlim(-0.05, 1)
ylim(-1.5 ,1.5)

legend([r"$v_{\rm I}(t), R_{\rm K} = R_0$", r"$v_{\rm I}(t), R_{\rm K} = 0$"])

plt.savefig("zero_rhs_plot.pdf", bbox_inches='tight')

```

```python
generate_canvas( w = .4 * 165 / 24.5, a = 1.5 )
#ylim([-1.5, 1.5])

K = -1 
tau_0 = 200e-6 

tau_rhp = 100e-6 

def f(t):
    if t < 0: return 0
    return 5 * (1 - 1/2*exp(-2*t) - 1/sqrt(2) * cos(2*t - pi/4) )

def g(t):
    return 5 * (1 - 1/sqrt(2) * cos(2*t - pi/4) )


def h(t):
    if t < 0: return 0
    return 5 * (- 1/2*exp(-2*t) )


tmax = 10
t_vec = linspace(-tmax, tmax, 2000)

v1 = [ f(t) for t in t_vec ]
v2 = [ g(t) for t in t_vec ]
v3 = [ h(t) for t in t_vec ]


plot(t_vec, v1, 'k--', lw = 1.5)
plot(t_vec, v2, 'k-', lw = 1.5)
plot(t_vec, v3, 'k:', lw = 1.5)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$y\rm\,[cm]$")

legend([r"$y$", r"$y_{\rm ss}$", r"$y_{\rm t}$"], loc = "lower right")

xlim(-2, 10)
ylim(-5 ,10)


plt.savefig("cim_rez_razni_plot.pdf", bbox_inches='tight')

```
