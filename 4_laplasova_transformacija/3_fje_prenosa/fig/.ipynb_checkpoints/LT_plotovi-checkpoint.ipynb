```python
from pylab import *
import matplotlib as mpl
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif"
    })

plt.rc('text.latex', preamble=r'\usepackage[T1,T2A]{fontenc}\usepackage[utf8]{inputenc}\usepackage{cmsrb}\usepackage{amsfonts,amsmath}\usepackage[Symbolsmallscale]{upgreek}\usepackage{icomma}')

font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)


# Decimalni zarez
import locale
locale.setlocale(locale.LC_NUMERIC, "sr_RS.utf8")
plt.rcParams['axes.formatter.use_locale'] = True


def u(t):
    return (t > 0) * 1.0

def generate_canvas(w = 1.6, a = 1.618):
    fig = figure (figsize = (w, w/a), dpi = 200)
    xlim([-2,2])
    ylim([-2,2])
    minorticks_on()
    grid(b=True, which='major', color='0.45', linestyle='-')
    grid(b=True, which='minor', color='0.75', linestyle='--')

    return fig

def plot_dirac(x, A, lw = 1.5, c = 'k'):
    arrow(x, 0, 0, A, shape="full", color=c,linewidth=lw, head_width=0.2, head_length=0.3, length_includes_head=True, zorder=10)
    
    
```

```python
generate_canvas( w = 165 / 24.5, a = 4 )
ylim([-1.5, 1.5])

T = 20

def f(t):
    if t < 0: return 0
    if t < T: return 62.5*sqrt(  (1 - exp(-t/18)) )
    return 62.5 * sqrt ( ( exp( T/18 ) - 1 ) * exp(-t/18) )

def lin(t):
    if t < 0: return 0
    if t < T: return 62.5*(  (1 - exp(-t/18)) )
    return 62.5 * ( ( exp( T/18 ) - 1 ) * exp(-t/18) )


tmax = 80
t = linspace(0, tmax, 1000)

v = [ f(t) for t in t ]

xlim([0, tmax])
ylim([0, 65])

plot(t, v, 'k-', lw = 1.5)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$v\rm\,[km/h]$")

#plt.savefig("tram_plot.pdf", bbox_inches='tight')
```

```python
generate_canvas( w = .65 * 165 / 24.5, a = 2 )
#ylim([-1.5, 1.5])

K = -1 
tau_0 = 200e-6 

tau_rhp = 100e-6 


def f(t):
    if t < 0: return 0
    return K * (1 - exp(-t/tau_0))

def g(t):
    if t < 0: return 0
    return K + (1-K)*exp(-t/tau_rhp)


tmax = 1000e-6
t_vec = linspace(-tmax, tmax, 2000)

v1 = [ f(t) for t in t_vec ]
v2 = [ g(t) for t in t_vec ]


plot(t_vec/1e-3, v1, 'k-', lw = 1.5)
plot(t_vec/1e-3, v2, 'k--', lw = 1)
xlabel(r"$t\rm\,[ms]$")
ylabel(r"$v\rm\,[V]$")

xlim(-0.05, 1)
ylim(-1.5 ,1.5)

legend([r"$v_{\rm I}(t), R_{\rm K} = R_0$", r"$v_{\rm I}(t), R_{\rm K} = 0$"])

plt.savefig("zero_rhs_plot.pdf", bbox_inches='tight')

```

```python
generate_canvas( w = .4 * 165 / 24.5, a = 1.5 )
#ylim([-1.5, 1.5])

K = -1 
tau_0 = 200e-6 

tau_rhp = 100e-6 

def f(t):
    if t < 0: return 0
    return 5 * (1 - 1/2*exp(-2*t) - 1/sqrt(2) * cos(2*t - pi/4) )

def g(t):
    return 5 * (1 - 1/sqrt(2) * cos(2*t - pi/4) )


def h(t):
    if t < 0: return 0
    return 5 * (- 1/2*exp(-2*t) )


tmax = 10
t_vec = linspace(-tmax, tmax, 2000)

v1 = [ f(t) for t in t_vec ]
v2 = [ g(t) for t in t_vec ]
v3 = [ h(t) for t in t_vec ]


plot(t_vec, v1, 'k--', lw = 1.5)
plot(t_vec, v2, 'k-', lw = 1.5)
plot(t_vec, v3, 'k:', lw = 1.5)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$y\rm\,[cm]$")

legend([r"$y$", r"$y_{\rm ss}$", r"$y_{\rm t}$"], loc = "lower right")

xlim(-2, 10)
ylim(-5 ,10)


plt.savefig("cim_rez_razni_plot.pdf", bbox_inches='tight')

```

```python
generate_canvas( w = .4 * 165 / 24.5, a = 1.5 )
ylim([-1.5, 1.5])

tmax=30

def feps(t):
    if t < 0: return 0
    return exp(-0.31416*t)*(cos(0.727*t) - 0.603* sin(0.727*t) )

def fenv1(t):
    if t < 0: return 0
    return exp(-0.31416*t)

def fenv2(t):
    if t < 0: return 0
    return -exp(-0.31416*t)


t_vec = linspace(0, tmax, 1000)

v = [ feps(t) for t in t_vec ]
env1 = [ fenv1(t) for t in t_vec ]
env2 = [ fenv2(t) for t in t_vec ]

xlim([0, tmax])
ylim([-1, 1])

plot(t, v, 'k-', lw = 1.5)
plot(t, env1, 'k--', lw = 1)
plot(t, env2, 'k--', lw = 1)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$\uptheta/\uptheta_0$")

plt.savefig("teleskop_plot.pdf", bbox_inches='tight')
```

```python
generate_canvas( w = 1.5*.4 * 165 / 24.5, a = 2 )
ylim([-1.5, 1.5])

wk = sqrt(9.81)
tau = 15
w0 = sqrt(wk**2 - 1/(4*tau**2))
w = 1.05 * wk 
H0 = 1e-8 * 1000
VG = 9e3
sigma = 1/30

tmax = 100

def f(t):
    if t < 0: return 0
    return 2*H0*VG/pi * 1/(w0**2 - w**2) * (sin(w*t) - w/w0 * exp(-sigma*t)*sin(w0*t) )

dw = w0 - w

def fenv(t):
    if t < 0: return 0
    return 2*H0*VG/pi * 1/(2*w*dw) * (  1 - (1 - dw/w)*exp(-sigma*t)*cos( dw*t ) )


t_vec = linspace(0, tmax, 10000)

v = [ f(t)*180/pi for t in t_vec ]
venv = [ -fenv(t)*180/pi for t in t_vec ]

xlim([0, tmax])
ylim([-7, 7])

plot(t_vec, v, 'k-', lw = .75)
plot(t_vec, venv, 'k--', lw = 1)
xlabel(r"$t\rm\,[s]$")
ylabel(r"$\uptheta\rm\,[^{\circ}]$")
legend([r"$\uptheta$", r"$A$"])

plt.savefig("klatno_plot.pdf", bbox_inches='tight')
```

```python
generate_canvas( w = 1.5*.4 * 165 / 24.5, a = 2 )
#ylim([-1.5, 1.5])

wk = sqrt(9.81)
tau = 15
w0 = sqrt(wk**2 - 1/(4*tau**2))
w = 1.05 * wk 
H0 = 1e-8 * 1000
VG = 9e3
sigma = 1/30

def f(t):
    if t < 0: return 0
    return 1 - exp(-50.0*t) * cos(t*1e3)

def fenv_1(t):
    if t < 0: return 0
    return 1 + exp(-50.0*t)

def fenv_2(t):
    if t < 0: return 0
    return 1 - exp(-50.0*t)


tmax =  2.5/50


t_vec = linspace(-tmax/20, tmax, 1000)

v = [ f(t) for t in t_vec ]
venv1 = [ fenv_1(t) for t in t_vec ]
venv2 = [ fenv_2(t) for t in t_vec ]

plot(t_vec*1e3, v, 'k-', lw = 1)
plot(t_vec*1e3, venv1, 'k--', lw = 0.75)
plot(t_vec*1e3, venv2, 'k--', lw = 0.75)
xlabel(r"$t\rm\,[ms]$")
ylabel(r"$f(t)$")

xlim(-tmax*1e3/20, tmax*1e3)
ylim(-.1, 2)

plt.savefig("Q_odziv.pdf", bbox_inches='tight')
```

```python
w = 1.5 * 165 / 24.5 / 2
a = 2 * 4 / 2

#fig = figure (figsize = (w, w/a), dpi = 200)
fig, ax = plt.subplots(3, 2, figsize = (w, w/a * 4), dpi = 200)
xlim([-2,2])
ylim([-2,2])
minorticks_on()
#grid(b=True, which='major', color='0.45', linestyle='-')
#grid(b=True, which='minor', color='0.75', linestyle='--')

Q_vec = [ 1, 1,    10,    10   ,  100, 100 ]
wNvec = [ 1e3, 2e3, 1e3, 2e3,  1e3, 2e3 ]

for Q,wN,m in zip(Q_vec, wNvec, range(2*4)):

    sigma = -wN/(2*Q)
    w0 = wN*sqrt(1 - 1/(4*Q**2))
    phi = arctan(sqrt(4*Q**2 - 1))

    def f(t):
        if t < 0: return 0
        return 1 - 1/sqrt(1 - 1/(4*Q**2))* exp(sigma*t) * sin(w0*t + phi )

    def fenv_1(t):
        if t < 0: return 0
        return 1 + exp(sigma*t)

    def fenv_2(t):
        if t < 0: return 0
        return 1 - exp(sigma*t)


    tmax =  2.5/50

    t_vec = linspace(0, tmax, 1000)

    v = [ f(t) for t in t_vec ]
    venv1 = [ fenv_1(t) for t in t_vec ]
    venv2 = [ fenv_2(t) for t in t_vec ]

    ax.flatten()[m].plot(t_vec*1e3, v, 'k-', lw = 1.25)
    ax.flatten()[m].plot(t_vec*1e3, venv1, 'k--', lw = 0.75)
    ax.flatten()[m].plot(t_vec*1e3, venv2, 'k--', lw = 0.75)
    ax.flatten()[m].set_xlabel(r"$t\rm\,[ms]$")
    ax.flatten()[m].set_ylabel(r"$f(t)$")
    ax.flatten()[m].set_title(r"$Q = " + str(Q) + r", \,\, \upomega_\mathrm{N} = " + str(wN/1000.0).replace('.', ',') + r"\rm\, [krad/s]$", fontsize=10)

    ax.flatten()[m].minorticks_on()
    ax.flatten()[m].set_xlim(0, tmax*1e3)
    ax.flatten()[m].set_ylim(0, 2)
    ax.flatten()[m].grid(b=True, which='major', color='0.45', linestyle='-')
    ax.flatten()[m].grid(b=True, which='minor', color='0.75', linestyle='--')

plt.tight_layout()
plt.savefig("Q_razliciti_odzivi.pdf", bbox_inches='tight')
```
